```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating the data frame
dataframe = pd.read_csv("Zomato-data-.csv")
print(dataframe.head())
```

```python
#Data Cleaning and Preparation
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
# Getting summary of the dataframe use df.info()
dataframe.info()
```

```python
# Value Count
dataframe['listed_in(type)'].value_counts()
```

```python
#Exploring Restaurant Types
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

```python
# Votes by Restaurant Type
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Type of restaurant', c='red', size=20)
plt.ylabel('Votes', c='red', size=20)
```

```python
# Identify the Most Voted Restaurant
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print('Restaurant(s) with the maximum votes:')
print(restaurant_with_max_votes)
```

```python
# Online Order Availability
sns.countplot(x=dataframe['online_order'])
```

```python
# Analyze Ratings
plt.hist(dataframe['rate'],bins=5)
plt.title('Ratings Distribution')
plt.show()
```

```python
# Approximate Cost for Couples
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x='approx_cost(for two people)', data=dataframe, hue='approx_cost(for two people)', palette='Set3', dodge=False, legend=False)
```

```python
# Ratings Comparison - Online vs Offline Orders
plt.figure(figsize = (6,5))
sns.boxplot(x='online_order', y='rate', hue='online_order', data=dataframe, palette={'Yes': 'skyblue', 'No': 'orange'}, dodge=False)
```

```python
# Order Mode Preferences by Restaurant Type
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```
